```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Function to extract important molecular descriptors
def calculate_descriptors(smiles):
    if isinstance(smiles, str) and smiles:
        mol = Chem.MolFromSmiles(smiles)
        if mol:
            descriptors = {
                'LogP': Descriptors.MolLogP(mol),  # Lipophilicity
                'MolecularWeight': Descriptors.MolWt(mol),  # Molecular Weight
                'HDonors': Descriptors.NumHDonors(mol),  # Number of Hydrogen Bond Donors
                'HAcceptors': Descriptors.NumHAcceptors(mol),  # Number of Hydrogen Bond Acceptors
                'TPSA': Descriptors.TPSA(mol)  # Topological Polar Surface Area
            }
            return descriptors
    return None

print(new_data['Smiles'].dtype)

# Drop rows where 'Smiles' is NaN or not a string
new_data = new_data[new_data['Smiles'].apply(lambda x: isinstance(x, str) and x.strip() != "")]

descriptor_data = new_data['Smiles'].apply(calculate_descriptors)

descriptor_df = pd.DataFrame(descriptor_data.tolist())

new_data = pd.concat([new_data.reset_index(drop=True), descriptor_df.reset_index(drop=True)], axis=1)

new_data = new_data.dropna(subset=['LogP', 'MolecularWeight', 'HDonors', 'HAcceptors', 'TPSA'])

new_data
```
